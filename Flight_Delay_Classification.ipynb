```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, dayofweek, dayofmonth

from pyspark.ml.feature import StringIndexer,  VectorAssembler
from pyspark.ml import Pipeline
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
    # .master("spark://172.17.176.1:7077") \
spark = SparkSession.builder \
    .appName("Flight Delay Prediction") \
    .getOrCreate()
```

```python
# Load dataset
data = spark.read.csv("./datasets/flight_data.csv", inferSchema=True, header=True)

# Show the schema of the data
data.printSchema()
```

## Preprocessing

### Filter rows where flights are cancelled or diverted

```python
data = data.filter((col("CANCELLED") == 0) & (col("DIVERTED") == 0))
```

### Making a combination of delays

```python
# Set delay threshold
delay_threshold = 15

# Create a "ReasonOfDelay" column
data = data.withColumn(
    "REASON_OF_DELAY",
    when(col("DELAY_DUE_CARRIER") >= delay_threshold, "CarrierDelay")
    .when(col("DELAY_DUE_WEATHER") >= delay_threshold, "WeatherDelay")
    .when(col("DELAY_DUE_NAS") >= delay_threshold, "NASDelay")
    .when(col("DELAY_DUE_SECURITY") >= delay_threshold, "SecurityDelay")
    .when(col("DELAY_DUE_LATE_AIRCRAFT") >= delay_threshold, "LateAircraftDelay")
    .otherwise("No Delay")
)
```

### Extract day of the week from flight date

```python
data = data.withColumn("DAY_OF_WEEK", dayofweek(col("FL_DATE")))
data = data.withColumn("DAY_OF_MONTH", dayofmonth(col("FL_DATE")))
```

### Feature Selection

```python
columns_to_drop = ['AIRLINE_DOT', 'AIRLINE_CODE', 'DOT_CODE', 'ORIGIN_CITY', 'DEST_CITY', 'DEP_TIME', 'DEP_DELAY','TAXI_OUT', \
                   'WHEELS_OFF', 'WHEELS_ON', 'TAXI_IN', 'ARR_TIME', 'ARR_DELAY', 'CANCELLED', 'CANCELLATION_CODE', 'DIVERTED','CRS_ELAPSED_TIME',\
                   'ELAPSED_TIME', 'AIR_TIME', 'DELAY_DUE_CARRIER','DELAY_DUE_WEATHER','DELAY_DUE_NAS','DELAY_DUE_SECURITY','DELAY_DUE_LATE_AIRCRAFT']

data = data.drop(*columns_to_drop).dropna()
```

```python
data.show()
```

```python
# Convert to Pandas for visualization
data_pandas = data.toPandas()
```

### Encode categorical variables: AIRLINE, ORIGIN, DEST, REASON_OF_DELAY

```python
indexer_airline = StringIndexer(inputCol="AIRLINE", outputCol="AIRLINE_index")
indexer_origin = StringIndexer(inputCol="ORIGIN", outputCol="ORIGIN_index")
indexer_dest = StringIndexer(inputCol="DEST", outputCol="DEST_index")
indexer_reason = StringIndexer(inputCol="REASON_OF_DELAY", outputCol="REASON_OF_DELAY_index")
```

### Making a pipeline

```python
# Assemble features into a feature vector
assembler = VectorAssembler(inputCols=["AIRLINE_index", "ORIGIN_index", "DEST_index", 
                                       "CRS_DEP_TIME", "CRS_ARR_TIME", "DISTANCE", "DAY_OF_WEEK", "DAY_OF_MONTH"],
                            outputCol="features")

# Define the pipeline
pipeline = Pipeline(stages=[indexer_airline, indexer_origin, indexer_dest, indexer_reason, assembler])
pipeline_model = pipeline.fit(data)
processed_data = pipeline_model.transform(data)
processed_data.select("features", "REASON_OF_DELAY_index").show(5)
```

### Split data into training and testing sets (70% for training, 30% for testing)

```python
train_data, test_data = processed_data.randomSplit([0.7, 0.3], seed=42)
```

## Model Training

### Logistic Regression

```python
log_reg = LogisticRegression(
    labelCol="REASON_OF_DELAY_index",
    featuresCol="features",
    regParam=0.1,            # Regularization parameter
    elasticNetParam=0.8,     # ElasticNet parameter
    maxIter=100             # Number of iterations
)

# Train the model
log_reg_model = log_reg.fit(train_data)
```

### Random Forest

```python
rf = RandomForestClassifier(
    labelCol="REASON_OF_DELAY_index",
    featuresCol="features",
    numTrees=10,             # Number of trees in the forest
    maxDepth=5,              # Maximum depth of each tree
    maxBins=380               # Increase maxBins to handle large categorical features
)
    
# Train the model
rf_model = rf.fit(train_data)
```

## Evaluation

### Eval function

```python
def eval(predictions):
    evaluator_accuracy = MulticlassClassificationEvaluator(labelCol="REASON_OF_DELAY_index",predictionCol="prediction", metricName="accuracy")
    evaluator_precision = MulticlassClassificationEvaluator(labelCol="REASON_OF_DELAY_index",predictionCol="prediction", metricName="weightedPrecision")
    evaluator_recall = MulticlassClassificationEvaluator(labelCol="REASON_OF_DELAY_index",predictionCol="prediction", metricName="weightedRecall")
    evaluator_f1 = MulticlassClassificationEvaluator(labelCol="REASON_OF_DELAY_index",predictionCol="prediction", metricName="f1")

    accuracy = evaluator_accuracy.evaluate(predictions)
    precision = evaluator_precision.evaluate(predictions)
    recall = evaluator_recall.evaluate(predictions)
    f1 = evaluator_f1.evaluate(predictions)

    return accuracy, f1, precision, recall
```

### Logistic Regression

```python
# Make predictions
log_reg_predictions = log_reg_model.transform(test_data)

log_reg_accuracy, log_reg_f1, log_reg_precision, log_reg_recall =  eval(log_reg_predictions)

# Print the evaluation metrics
print(f"Logistic Regression Accuracy: {log_reg_accuracy:.4f}")
print(f"Logistic Regression Precision: {log_reg_precision:.4f}")
print(f"Logistic Regression Recall: {log_reg_recall:.4f}")
print(f"Logistic Regression F1-Score: {log_reg_f1:.4f}")
```

### Random Forest

```python
 # Make predictions
rf_predictions = rf_model.transform(test_data)
    
# Evaluate model
rf_accuracy, rf_f1, rf_precision, rf_recall =  eval(rf_predictions)
    
print(f"Random Forest Accuracy: {rf_accuracy:.4f}")
print(f"Random Forest Precision: {rf_precision:.4f}")
print(f"Random Forest Recall: {rf_recall:.4f}")
print(f"Random Forest F1-Score: {rf_f1:.4f}")
```

### Comparing the models

```python
import pandas as pd

# Define the model names and evaluation metrics
model_names = ['Logistic Regression', 'Random Forest']
accuracies = [log_reg_accuracy, rf_accuracy]
precisions = [log_reg_precision, rf_precision]
recalls = [log_reg_recall, rf_recall]
f1_scores = [log_reg_f1, rf_f1]

# Create a dictionary of the model performance metrics
model_comparison_dict = {
    'Model': model_names,
    'Accuracy': accuracies,
    'Precision': precisions,
    'Recall': recalls,
    'F1-Score': f1_scores
}

# Create a DataFrame from the dictionary
model_comparison_df = pd.DataFrame(model_comparison_dict)

# Display the comparison DataFrame
print(model_comparison_df)
```
